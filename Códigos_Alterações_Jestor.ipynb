```python
from google.colab import userdata
import json
import requests
import time


id_tabela_main = "w5s8rut7rnwph47t5hobl"
id_tabela_aux = "6_71f3xvx_yafybyz8o0m"


# Montando Autenticação
full_token_jestor = userdata.get('token_jestor')

# Montando nome do campo de id
info_registro = "id_" + id_tabela_aux

# Buscando registro pelo título para encontrar ID
url_list = "https://loft.api.jestor.com/object/list"
url_create = "https://loft.api.jestor.com/object/create"
url_busca = "https://loft.api.jestor.com/object/list"
url_delete = "https://loft.api.jestor.com/object/delete"
url_update = "https://loft.api.jestor.com/object/update"


# Montando obj da tabela aux
#================================================================================================
payload_list = {
    "object_type": id_tabela_aux,
    "page": 1,
    "size": 1000000
}

headers = {
    "accept": "application/json",
    "content-type": "application/json",
    "Authorization": full_token_jestor
}

response = requests.post(url_list, json=payload_list, headers=headers)
# print(response.text)

response_json = json.loads(response.text)
results = response_json.get("data", {}).get("items", [])


for result in results:
  # print(result)
  obj = {
      "name": result["name"],
      "nome_completo": result["nome_completo"],
      "documento": result["documento"],
      "forma_de_pagamento": result["forma_de_pagamento"],
      "tipo_de_chave": result["tipo_de_chave"],
      "pix": result["pix"],
      "banco": result["banco"],
      "agencia": result["agencia"],
      "conta_com_digito": result["conta_com_digito"],
  }
  doc = result["documento"]
  id_aux = result["id_" + id_tabela_aux]

  #================================================================================================



  # Pesquisando caso na tabela principal
  #================================================================================================

  payload_busca = {
      "object_type": id_tabela_main,
      "filters": [
          {
              "field": "documento",
              "operator": "==",
              "value": doc
          }
      ]
  }

  response_busca = requests.post(url_busca, json=payload_busca, headers=headers)
  response_json_busca = json.loads(response_busca.text)
  results_busca = response_json_busca.get("data", {}).get("items", [])


  if(len(results_busca) > 0):

    record = results_busca[0]
    id_jestor = record["id_" + id_tabela_main]
    id_tab_update = "id_" + id_tabela_main
    print(f"Updating id={id_jestor} na tabela {id_tabela_main}")

  #================================================================================================
  # Update na tabela principal
    payload_update = {
        "object_type": id_tabela_main,
        "data": {
            id_tab_update : id_jestor,
            "nome_completo": result["nome_completo"],
            "documento": result["documento"],
            "forma_de_pagamento": result["forma_de_pagamento"],
            "tipo_de_chave": result["tipo_de_chave"],
            "pix": result["pix"],
            "banco": result["banco"],
            "agencia": result["agencia"],
            "conta_com_digito": result["conta_com_digito"],
        }
    }
    response_update = requests.post(url_update, json=payload_update, headers=headers)


  else:
    print("Creating...")

  #================================================================================================
  # Criando registro na tabela principal
    payload_create = {
        "object_type": id_tabela_main,
        "data": obj
    }
    response_create = requests.post(url_create, json=payload_create, headers=headers)




  #================================================================================================
  # Apagando registro na tabela auxiliar
  print(f"Excluindo id {id_aux} na tabela {id_tabela_aux}")
  payload_delete = {
      "object_type": id_tabela_aux,
      "id": id_aux
  }
  response_delete = requests.post(url_delete, json=payload_delete, headers=headers)

  time.sleep(1)
```
